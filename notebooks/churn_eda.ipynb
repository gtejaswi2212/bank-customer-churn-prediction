# Bank Customer Churn — Exploratory Data Analysis

This notebook explores the analytical base table used for churn prediction: distributions, correlations, and churn rates by segment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

ROOT = Path("..")
df = pd.read_csv(ROOT / "analytical_base_table.csv")
df.info()
```

```python
# Churn distribution
df["Exited"].value_counts(normalize=True).plot(kind="bar")
plt.title("Churn Distribution")
plt.xticks([0, 1], ["Retained", "Churned"])
plt.ylabel("Proportion")
plt.tight_layout()
plt.show()
```

```python
# Numeric features vs churn
numeric = ["CreditScore", "Age", "Tenure", "Balance", "NumOfProducts", "EstimatedSalary"]
df[numeric + ["Exited"]].groupby("Exited").mean().T.plot(kind="bar", figsize=(10, 4))
plt.legend(["Retained", "Churned"])
plt.xticks(rotation=45, ha="right")
plt.title("Mean feature values by churn")
plt.tight_layout()
plt.show()
```

```python
# Correlation with target
corr = df.select_dtypes(include=['number']).corr()["Exited"].drop("Exited", errors='ignore').sort_values(ascending=True)
corr.plot(kind="barh")
plt.title("Correlation with Churn")
plt.tight_layout()
plt.show()
```
